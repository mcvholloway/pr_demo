```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, np.pi*2, 500)
y = np.sin(x)**2
```

```python
plt.plot(x,y);
```

```python
x = np.linspace(0, 1, 500)
y = np.arcsin(x)**2
```

```python
plt.plot(x,y);
```

```python
x = np.linspace(0, np.pi*2, 500)
y = np.tan(x)**2
```

```python
plt.plot(x,y);
```
